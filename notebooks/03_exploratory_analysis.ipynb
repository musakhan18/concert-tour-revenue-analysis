## 3. Univariate Analysis

```python
# Distribution of numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col].dropna(), kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
# Bar plots for categorical columns
categorical_cols = df.select_dtypes(include="object").columns

for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    df[col].value_counts().head(10).plot(kind="bar")
    plt.title(f"Top Categories in {col}")
    plt.show()
```

## 4. Bivariate Analysis

```python
# Scatter plots between revenue and other numeric features
target_col = "Actual gross" 

for col in numeric_cols:
    if col != target_col:
        plt.figure(figsize=(6, 4))
        sns.scatterplot(data=df, x=col, y=target_col)
        plt.title(f"{col} vs {target_col}")
        plt.show()
```

```python
# Grouped bar chart example
if "Artist" in df.columns and target_col in df.columns:
    artist_revenue = df.groupby("Artist")[target_col].mean().sort_values(ascending=False).head(10)
    artist_revenue.plot(kind="bar", figsize=(8, 5), color="teal")
    plt.title("Top 10 Artists by Average Gross Revenue")
    plt.show()
```

## 5. Correlation Analysis

```python
# Correlation matrix
corr = df.corr(numeric_only=True)

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Strongest correlations with target
corr_target = corr[target_col].sort_values(ascending=False)
print("Top correlations with target column:")
print(corr_target)
```

## 6. Outlier Detection

```python
# Boxplots for numeric columns
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df[col])
    plt.title(f"Outlier Check for {col}")
    plt.show()
```
